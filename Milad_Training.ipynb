```python
# Taking the feature engineering analyis csv file into my directory
import pandas
feature_eng_data = pandas.read_csv('data/new_sm_train.csv')
```

```python
# KNN Model
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# Select columns to scale (numerical columns)

# numerical_columns = ['latitude_x', 'longitude_x', 'latitude_y', 'longitude_y', 'delivery_charge', 
#                      'serving_distance', 'prepration_time', 'discount_percentage', 'vendor_rating', 
#                      'distance', 'distance_scaled', 'ordered_vendor_tag', 'matching_tags_count']

# # Create a scaler object
# scaler = StandardScaler()

# # Apply scaling
# feature_eng_data[numerical_columns] = scaler.fit_transform(feature_eng_data[numerical_columns])


# Trying with all features and dropping customer ID and vendor ID
# Uncooment to apply scaling
numerical_columns = ['latitude_x', 'longitude_x', 'latitude_y',
       'longitude_y', 'vendor_category_id', 'delivery_charge',
       'serving_distance', 'prepration_time', 'discount_percentage',
       'vendor_rating', 'ordered_before', 'vendor_average',
       'customer_average', 'is_favorite', 'times_ordered', 'is_rated',
       'customer_rating', 'distance', 'distance_scaled', 'ordered_vendor_tag',
       'matching_tags_count']

# Create a scaler object
scaler = StandardScaler()

# Apply scaling
feature_eng_data[numerical_columns] = scaler.fit_transform(feature_eng_data[numerical_columns])
```

```python
# Define the features (X) - drop the customer_id and vendor_id columns from the feature set
X = feature_eng_data.drop(['customer_id', 'vendor_id', 'target'], axis=1)

# Define the target variables for customer and vendor
y_customer = feature_eng_data['customer_id']
y_vendor = feature_eng_data['vendor_id']
y_target = feature_eng_data['target']
```

```python
# # Create an imputer to fill missing values with the mean 
# imputer = SimpleImputer(strategy='mean')  

# # Apply the imputer to the features (X)
# X_imputed = imputer.fit_transform(X)

# # Now proceed with the model training
# X_train_customer, X_test_customer, y_train_customer, y_test_customer = train_test_split(X_imputed, y_customer, test_size=0.2, random_state=42)
# X_train_vendor, X_test_vendor, y_train_vendor, y_test_vendor = train_test_split(X_imputed, y_vendor, test_size=0.2, random_state=42)


# Fixed missing values, now proceed with the model training
X_tr_customer, X_te_customer, y_tr_customer, y_te_customer = train_test_split(X, y_customer, test_size=0.2, random_state=42)
X_tr_vendor, X_te_vendor, y_tr_vendor, y_te_vendor = train_test_split(X, y_vendor, test_size=0.2, random_state=42)
X_tr_target, X_te_target, y_tr_target, y_te_target = train_test_split(X, y_target, test_size=0.2, random_state=42)
```

```python
# Initialize and train KNN models
knn_customer = KNeighborsClassifier(n_neighbors=5)
knn_customer.fit(X_tr_customer, y_tr_customer)

knn_vendor = KNeighborsClassifier(n_neighbors=5)
knn_vendor.fit(X_tr_vendor, y_tr_vendor)

# Add target KNN model
knn_target = KNeighborsClassifier(n_neighbors=5)
knn_target.fit(X_tr_target, y_tr_target)
```

```python
from sklearn.metrics import classification_report

# Evaluate customer_id model
y_pred_customer = knn_customer.predict(X_te_customer)
accuracy_customer = accuracy_score(y_te_customer, y_pred_customer)
print(f"Customer Model Accuracy: {accuracy_customer * 100:.2f}%")

# Evaluate vendor_id model
y_pred_vendor = knn_vendor.predict(X_te_vendor)
accuracy_vendor = accuracy_score(y_te_vendor, y_pred_vendor)
print(f"Vendor Model Accuracy: {accuracy_vendor * 100:.2f}%")

# Evaluate target model
y_pred_target = knn_target.predict(X_te_target)
accuracy_target = accuracy_score(y_te_target, y_pred_target)
print(f"Target Model Accuracy: {accuracy_target * 100:.2f}%")

# Including classification report for target model
print(classification_report(y_te_target, y_pred_target))
```
